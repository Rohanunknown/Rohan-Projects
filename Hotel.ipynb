# OUTLIERS TREATMENT

```python
df.columns
```

```python
# Function to handle outliers by replacing them with the median
def handle_outliers(df, column):
    median = df[column].median()
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)

handle_outliers(df,'lead_time')
handle_outliers(df,'avg_price_per_room')
handle_outliers(df,'no_of_week_nights')

print("\nModified dataset:")
print(df.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Define the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Define the hyperparameters to tune
param_grid = {
    'n_estimators': [100, 200, 300],        # Number of trees in the forest
    'max_depth': [10, 20, 30],              # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],        # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],          # Minimum samples required at a leaf node
    'bootstrap': [True, False]              # Whether to use bootstrap samples
}

# Use GridSearchCV to tune hyperparameters
rf_grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
rf_grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best Parameters: ", rf_grid_search.best_params_)
print("Best Accuracy: ", rf_grid_search.best_score_)

# Predictions on the test set
y_pred_rf = rf_grid_search.predict(X_test)

# Evaluate the model
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

```python
# Accuracy on training data
train_accuracy = rf_grid_search.score(X_train, y_train)
print("Training Accuracy: ", train_accuracy)

# Accuracy on test data
test_accuracy = rf_grid_search.score(X_test, y_test)
print("Test Accuracy: ", test_accuracy)
```

## RANDOM FOREST MODEL IS OVERFITTING
SO WE HAVE TO CONTROL OVERFITTING
SO WE HAVE TO DO SOME CHANGE IN HYPERPARAMETER INCRESING NUMBER OF TREES,ALOW DEEP TREES,INCREASE MINIMUM SAMPLES,USING MAX FEATURES,USING MAXIMUM LEAF NODES

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Define the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Define a more detailed hyperparameter grid
param_grid = {
    'n_estimators': [200, 300, 400],           # More trees in the forest
    'max_depth': [10, 20, 30, None],           # Limit or allow deep trees
    'min_samples_split': [5, 10],              # Increase minimum samples to split a node
    'min_samples_leaf': [2, 4, 6],             # Increase minimum samples at leaf nodes
    'max_features': ['sqrt', 'log2'],          # Limit number of features at each split
    'bootstrap': [True],                       # Use bootstrap sampling (recommended)
    'max_leaf_nodes': [None, 50, 100]          # Limit the number of leaf nodes to prevent overfitting
}

# Initialize GridSearchCV
rf_grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)

# Fit the model
rf_grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters: ", rf_grid_search.best_params_)
print("Best Accuracy: ", rf_grid_search.best_score_)

# Use the best model to predict on the test set
best_rf = rf_grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)

# Evaluate performance
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

```python
# Accuracy on training data
train_accuracy = rf_grid_search.score(X_train, y_train)
print("Training Accuracy: ", train_accuracy)

# Accuracy on test data
test_accuracy = rf_grid_search.score(X_test, y_test)
print("Test Accuracy: ", test_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Define the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Define a new hyperparameter grid with stricter limits on max_depth
param_grid = {
    'n_estimators': [200, 300, 400],
    'max_depth': [5, 10, 15],               # Stricter limits on max depth
    'min_samples_split': [5, 10],
    'min_samples_leaf': [2, 4, 6],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True],
    'max_leaf_nodes': [None, 50, 100]
}

# Initialize GridSearchCV
rf_grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)

# Fit the model
rf_grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters: ", rf_grid_search.best_params_)
print("Best Accuracy: ", rf_grid_search.best_score_)

# Use the best model to predict on the test set
best_rf = rf_grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)

# Evaluate performance
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))

# Accuracy on training data
train_accuracy = rf_grid_search.score(X_train, y_train)
print("Training Accuracy: ", train_accuracy)

# Accuracy on test data
test_accuracy = rf_grid_search.score(X_test, y_test)
print("Test Accuracy: ", test_accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Predict probabilities instead of labels
y_probs_rf = best_rf.predict_proba(X_test)[:, 1]  # Get the probability for the positive class

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs_rf)

# Calculate AUC score
auc_score = roc_auc_score(y_test, y_probs_rf)
print("ROC-AUC Score: ", auc_score)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='ROC Curve (area = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line for random guessing
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs_rf)

# Calculate the average precision score
average_precision = average_precision_score(y_test, y_probs_rf)
print("Average Precision Score: ", average_precision)

# Plot the precision-recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', label='Precision-Recall Curve (AP = %0.2f)' % average_precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.show()
```
