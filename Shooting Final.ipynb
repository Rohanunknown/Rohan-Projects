# DATA CLEANING

#### Convert DATE AND TIME
OCCUR_DATE and OCCUR_TIME to a single datetime column.
Drop the original OCCUR_DATE and OCCUR_TIME columns.

```python
df['OCCUR_DATETIME'] = pd.to_datetime(df['OCCUR_DATE'] + ' ' + df['OCCUR_TIME'])

# Drop the original OCCUR_DATE and OCCUR_TIME columns
df_cleaned = df.drop(['OCCUR_DATE', 'OCCUR_TIME'], axis=1)
```

```python
df_cleaned.head()
```

```python
# Convert OCCUR_DATETIME to datetime format 
df_cleaned['OCCUR_DATETIME'] = pd.to_datetime(df_cleaned['OCCUR_DATETIME'], errors='coerce')

# Extract relevant datetime features
df_cleaned['OCCUR_YEAR'] = df_cleaned['OCCUR_DATETIME'].dt.year
df_cleaned['OCCUR_MONTH'] = df_cleaned['OCCUR_DATETIME'].dt.month
df_cleaned['OCCUR_DAY'] = df_cleaned['OCCUR_DATETIME'].dt.day
df_cleaned['OCCUR_HOUR'] = df_cleaned['OCCUR_DATETIME'].dt.hour

# Drop the original OCCUR_DATETIME column
df_cleaned.drop(columns=['OCCUR_DATETIME'], inplace=True)

# Check the updated data
print(df_cleaned.head())
```

##### Drop columns with significant missing data: LOC_OF_OCCUR_DESC, LOC_CLASSFCTN_DESC

```python
df_cleaned = df_cleaned.drop(['LOC_OF_OCCUR_DESC', 'LOC_CLASSFCTN_DESC'], axis=1)
```

#### Drop unnecessary columns: INCIDENT_KEY and Lon_Lat

```python
df_cleaned = df_cleaned.drop(['INCIDENT_KEY', 'Lon_Lat'], axis=1)
```

```python
df_cleaned['PERP_AGE_GROUP'].fillna('Unknown', inplace=True)
df_cleaned['PERP_SEX'].fillna('Unknown', inplace=True)
df_cleaned['PERP_RACE'].fillna('Unknown', inplace=True)
df_cleaned['LOCATION_DESC'].fillna('Unknown', inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize the data (scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Increase max_iter and use scaling (approach 1 & 2)
model_scaled = LogisticRegression(max_iter=2000)
model_scaled.fit(X_train_scaled, y_train)

# Predict and evaluate the model
y_pred_scaled = model_scaled.predict(X_test_scaled)

print("Confusion Matrix (Scaled Logistic Regression):")
print(confusion_matrix(y_test, y_pred_scaled))
print("\nClassification Report (Scaled Logistic Regression):")
print(classification_report(y_test, y_pred_scaled))

# Try a different solver (approach 3)
model_saga = LogisticRegression(solver='saga', max_iter=2000)
model_saga.fit(X_train_scaled, y_train)

# Predict and evaluate with the 'saga' solver
y_pred_saga = model_saga.predict(X_test_scaled)

print("Confusion Matrix (Logistic Regression with 'saga' solver):")
print(confusion_matrix(y_test, y_pred_saga))
print("\nClassification Report (Logistic Regression with 'saga' solver):")
print(classification_report(y_test, y_pred_saga))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.pipeline import Pipeline

# Create the pipeline with SMOTE and the classifier
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('hgb', HistGradientBoostingClassifier(random_state=42))
])

# Define the hyperparameter space to search
param_distributions = {
    'hgb__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'hgb__max_depth': [3, 5, 7],
    'hgb__min_samples_leaf': [10, 20, 50],
    'hgb__max_leaf_nodes': [10, 20, 50, 100],
    'hgb__max_iter': [100, 200, 300]
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    pipeline, 
    param_distributions, 
    n_iter=20,  # Number of parameter settings sampled
    scoring='f1_macro',  # Focus on F1 macro score to balance classes
    cv=3,  # 3-fold cross-validation
    verbose=2, 
    n_jobs=-1,  # Use all processors
    random_state=42
)

# Fit the RandomizedSearchCV to the training data
random_search.fit(X_train, y_train)

# Evaluate the best model
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

# Print best hyperparameters
print("Best hyperparameters:", random_search.best_params_)

# Print confusion matrix and classification report
print("Confusion Matrix (Tuned Model):")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report (Tuned Model):")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.combine import SMOTEENN
import re  # For sanitizing column names

# Assuming X_train, X_test, y_train, y_test are already defined

# Sanitize feature names by removing any invalid characters
def sanitize_column_names(df):
    df.columns = [re.sub(r'[\[\]<>,]', '_', str(col)) for col in df.columns]
    return df

# Apply sanitization to X_train and X_test
X_train = sanitize_column_names(X_train)
X_test = sanitize_column_names(X_test)

# Apply SMOTEENN for handling class imbalance
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)

# Ensure feature names of resampled data are strings and sanitized
X_resampled = pd.DataFrame(X_resampled, columns=[str(col) for col in X_train.columns])

# Initialize Gradient Boosting model
gb_model = GradientBoostingClassifier(random_state=42)

# Define the hyperparameter grid for GridSearchCV
param_grid = {
    'n_estimators': [500],
    'learning_rate': [0.05],
    'max_depth': [5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'subsample': [0.8],
    'max_features': ['sqrt', 'log2'],
}

# Perform GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, scoring='f1', cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_resampled, y_resampled)

# Best hyperparameters
print("Best Gradient Boosting hyperparameters:", grid_search.best_params_)

# Predict on the test set with the best model
best_gb_model = grid_search.best_estimator_
y_pred_proba = best_gb_model.predict_proba(X_test)[:, 1]

# Choose a custom threshold
threshold = 0.35  # Adjust this based on Precision-Recall trade-off
y_pred = (y_pred_proba >= threshold).astype(int)

# Confusion matrix and classification report
print("Confusion Matrix (SMOTEENN + Tuned Gradient Boosting with Threshold):")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report (SMOTEENN + Tuned Gradient Boosting with Threshold):")
print(classification_report(y_test, y_pred))

# Test with different thresholds and print results
thresholds = [0.25, 0.26, 0.27, 0.28, 0.29, 0.3, 0.31, 0.32, 0.33, 0.34, 0.35, 0.36]
for threshold in thresholds:
    y_pred = (y_pred_proba >= threshold).astype(int)
    print(f"\nThreshold: {threshold}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.combine import SMOTEENN
from sklearn.metrics import classification_report, confusion_matrix
import re

# Assuming X_train, X_test, y_train, y_test are already defined

# Sanitize feature names
def sanitize_column_names(df):
    df.columns = [re.sub(r'[\[\]<>,]', '_', str(col)) for col in df.columns]
    return df

X_train = sanitize_column_names(X_train)
X_test = sanitize_column_names(X_test)

# Apply SMOTEENN for handling class imbalance
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)

# Ensure feature names of resampled data are strings and sanitized
X_resampled = pd.DataFrame(X_resampled, columns=[str(col) for col in X_train.columns])

# Initialize the best Gradient Boosting model (based on GridSearchCV results)
best_gb_model = GradientBoostingClassifier(
    learning_rate=0.05,
    max_depth=5,
    max_features='log2',
    min_samples_leaf=1,
    min_samples_split=2,
    n_estimators=500,
    subsample=0.8,
    random_state=42
)

# Train the model on the resampled data
best_gb_model.fit(X_resampled, y_resampled)

# Predict probabilities on the test set
y_pred_proba = best_gb_model.predict_proba(X_test)[:, 1]

# Finalize the threshold
threshold = 0.35
y_pred = (y_pred_proba >= threshold).astype(int)

# Output the final performance
print("Confusion Matrix (Final Model with Threshold 0.35):")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report (Final Model with Threshold 0.35):")
print(classification_report(y_test, y_pred))

# Save the model for future use
import joblib
joblib.dump(best_gb_model, 'finalized_gradient_boosting_model.pkl')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc, confusion_matrix, roc_auc_score
import numpy as np


# 3. ROC Curve
def plot_roc_curve(y_test, y_pred_proba):
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()

# Call the ROC curve plot
plot_roc_curve(y_test, y_pred_proba)
```

# Explanation:
The diagonal gray line represents a random classifier (i.e., a model that makes random guesses), which would have an AUC of 0.5.
The blue line is your model's performance. The fact that it is above the diagonal means that your model performs better than random guessing.

Area under the ROC Curve (AUC = 0.66):
The AUC value of 0.66 indicates the model is moderately effective at distinguishing between fatal and non-fatal incidents.
A perfect classifier would have an AUC of 1.0, while a classifier that performs no better than random guessing would have an AUC of 0.5.
